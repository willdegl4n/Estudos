**Contando quantas vezes um elemento aparece:**

```python
# Recebe um nome do usuário
nome = input('Digite um nome: ')

# Recebe uma letra e a transforma em minúscula, com validação
'''
    Se o que for inserido for um único elemento e for uma letra, segue o código.
    Caso não for, o código exige uma entrada válida para continuar.
'''
while True:
    letra = input(f'Escolha uma letra do nome "{nome}" para descobrir a \
frequência: ').lower()
    if len(letra) == 1 and letra.isalpha():
        break
    else:
        print("Por favor, insira uma única letra válida.")

# Transforma o nome para minúsculo e conta a frequência da letra informada
frequencia = nome.lower().count(letra)

# Mostra a frequência que a letra informada aparece
print(f'Quantidade de letras "{letra}" no nome {nome}: {frequencia}')
```

**Opções e Formas de Verificação:**

```python
'''
O método .isalpha() é um método de string que verifica se todos os caracteres
em uma string são letras do alfabeto. Se todos os caracteres forem letras (A-Z
ou a-z) e a string não estiver vazia, .isalpha() retorna True. Caso contrário,
retorna False.
'''
print("abc".isalpha())   # True
print("abc123".isalpha()) # False
print("".isalpha())       # False
print("abc!".isalpha())   # False
print("Árvore".isalpha()) # True (considera caracteres acentuados como letras)


'''
O método .isdigit() verifica se todos os caracteres da string são dígitos (0-9).
'''
print("123".isdigit())  # True
print("123a".isdigit()) # False
print("".isdigit())     # False


'''
O método .isalnum() verifica se todos os caracteres da string são alfanuméricos
(letras e/ou números).
'''
print("abc123".isalnum()) # True
print("abc!".isalnum())   # False
print("123".isalnum())    # True
print("abc".isalnum())    # True
print("".isalnum())       # False


'''
O método .isspace() verifica se todos os caracteres da string são espaços em
branco (espaços, tabulações, etc.).
'''
print("   ".isspace()) # True
print("abc".isspace()) # False
print("\t\n".isspace()) # True


'''
.isupper() verifica se todos os caracteres da string são letras maiúsculas.
.islower() verifica se todos os caracteres da string são letras minúsculas.
'''
print("ABC".isupper())   # True
print("abc".isupper())   # False
print("ABC123".isupper()) # True (ignora números)
print("abc".islower())   # True
print("Abc".islower())   # False


'''
Utilizar len() para verificar o comprimento da string.
'''
letra = "a"
if len(letra) == 1:
    print("É uma única letra.")
else:
    print("Não é uma única letra.")
```

```python
def calcula_imc(peso, altura):
    """
    Calcula o Índice de Massa Corporal (IMC) dado o peso em kg e altura em metros.
    Retorna o IMC calculado.
    """
    return peso / (altura ** 2)

def obter_dado_valido(mensagem, tipo):
    """
    Solicita ao usuário um dado válido (peso ou altura) até que seja inserido corretamente.
    Retorna o valor convertido para float.
    """
    while True:
        try:
            dado = float(input(mensagem))
            if dado <= 0:
                raise ValueError(f"O {tipo} deve ser maior que zero.")
            return dado
        except ValueError as e:
            print(f"Erro: {e}. Por favor, digite um {tipo} válido.")

def main():
    peso = obter_dado_valido('Digite seu peso em kg (por exemplo, 70.5): ', 'peso')
    altura = obter_dado_valido('Digite sua altura em metros (por exemplo, 1.75): ', 'altura')

    imc = calcula_imc(peso, altura)

    print(f'\nSeu IMC é: {imc:.2f}')

if __name__ == "__main__":
    main()
```

**def calcula_imc(peso, altura):**

Aqui definimos uma função chamada calcula_imc que recebe dois parâmetros: peso e altura.

Essa função calcula o Índice de Massa Corporal (IMC) usando a fórmula padrão:

IMC = peso / altura elevada ao quadrado

**return peso / (altura ** 2):**

Retorna o valor calculado do IMC.
  
Aqui, altura ** 2 calcula o quadrado da altura, e então peso / (altura ** 2) realiza a divisão para obter o IMC.

**def obter_dado_valido(mensagem, tipo):**

Esta função obter_dado_valido é responsável por solicitar ao usuário um valor válido para peso ou altura.
   
Ela recebe dois parâmetros: mensagem, que é a mensagem que será exibida ao usuário para solicitar o dado, e tipo, que especifica se estamos pedindo o peso ou a altura.

Este é um loop infinito (while True:) que continuará solicitando entrada do usuário até que um valor válido seja fornecido.

**try:** e **except ValueError as e:**

**float(input(mensagem))**.

**if dado <= 0:**

**return dado:**

Se tudo estiver correto (ou seja, nenhum erro foi levantado e o valor é maior que zero), retornamos o valor convertido para float.

```python
# Função para carregar as opções do arquivo para o dicionário
def carregar_opcoes():
    try:
        with open('opcoes_ceu.txt', 'r') as arquivo:
            opcoes = arquivo.readlines()
            for linha in opcoes:
                opcao, mensagem = linha.strip().split(':')
                opcoes_ceu[opcao] = mensagem
    except FileNotFoundError:
        print('Arquivo de opções não encontrado. Usando opções padrão.')

# Função para salvar as opções do dicionário no arquivo
def salvar_opcoes():
    with open('opcoes_ceu.txt', 'w') as arquivo:
        for opcao, mensagem in opcoes_ceu.items():
            arquivo.write(f'{opcao}:{mensagem}\n')

# Define o dicionário inicial vazio
opcoes_ceu = {}

# Carrega as opções do arquivo
carregar_opcoes()

# Função para adicionar novas opções ao dicionário
def adicionar_opcao():
    nova_opcao = input('Digite uma nova condição do céu: ').lower()
    nova_mensagem = input('Digite a mensagem correspondente: ')
    opcoes_ceu[nova_opcao] = nova_mensagem
    salvar_opcoes()
    print('Opção adicionada com sucesso!')

# Loop para interação contínua com o usuário
while True:
    ceu = input('O céu está azul, cinza ou outra condição? (Digite "sair" para encerrar) ').lower()

    if ceu == 'sair':
        print('Encerrando o programa...')
        break
    elif ceu in opcoes_ceu:
        print(opcoes_ceu[ceu])
    else:
        print('Opção não reconhecida. Deseja adicionar uma nova opção? (sim/não)')
        resposta = input().lower()
        if resposta == 'sim':
            adicionar_opcao()
```

Explicação Linha por Linha do código abaixo:
